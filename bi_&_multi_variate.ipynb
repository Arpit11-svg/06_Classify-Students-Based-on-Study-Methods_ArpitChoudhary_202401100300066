```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips=sns.load_dataset('tips')
tips.sample(5)
```

```python
# NUMERICAL-NUMERICAL COLUMN
```

```python
# BIVARIATE ANALYSIS
sns.scatterplot(x=tips['total_bill'],y=tips['tip'],data=tips)
```

```python
# MULTIVARITE ANALYSIS
plt.figure(figsize=(16, 6))
sns.scatterplot(x=tips['total_bill'],y=tips['tip'],hue=tips['day'],style=tips['smoker'],size=tips['sex'],data=tips)
```

```python
titanic=sns.load_dataset('titanic')
titanic.sample(5)
```

```python
sns.barplot(data=titanic,x=titanic['pclass'],y=titanic['fare'],hue=titanic['sex'])
```

```python
sns.boxplot(data=titanic,x=titanic['sex'],y=titanic['age'],hue=titanic['survived'])
```

```python
 sns.distplot(titanic[titanic['survived']==0]['age'],kde=True,hist=False,color='red') #for alive
 sns.distplot(titanic[titanic['survived']==1]['age'],kde=True,hist=False,color='green') #for died
```

```python
# CATEGORICAL-CATEGORICAL COLUMN
```

```python
sns.heatmap(pd.crosstab(titanic['pclass'],titanic['survived']))
```

```python
sns.clustermap(pd.crosstab(titanic['sibsp'],titanic['survived']))
```
